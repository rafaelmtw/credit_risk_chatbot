```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "C:/Users/Samsung/Downloads/cleaned_german_credit_data.csv"  # replace with your file path
df = pd.read_csv(file_path)
# Iterate through each column and print unique responses
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Unique responses for '{column}':")
    print(unique_values)
    print(f"Count: {len(unique_values)}\n")
```

```python
df['Risk'] = df['Risk'].replace({'bad': 0, 'good': 1})
df['Saving accounts'] = df['Saving accounts'].replace({'little': 0, 'moderate': 1, 'quite rich': 2, 'rich': 3})
df['Checking account'] = df['Checking account'].replace({'little': 0, 'moderate': 1, 'rich': 2})
```

```python
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Unique responses for '{column}':")
    print(unique_values)
    print(f"Count: {len(unique_values)}\n")
```

```python
nominal_columns = df.select_dtypes(include=['object']).columns.tolist()

# One-hot encode the nominal columns
df_one_hot = pd.get_dummies(df, columns=nominal_columns, drop_first=True)

# Display the transformed DataFrame
print("\nOne-Hot Encoded DataFrame:")
print(df_one_hot)
```

```python
# Check the data types of each column
data_types = df_one_hot.dtypes

# Print the data types
print(data_types)
```

```python
risk = df_one_hot.pop('Risk')
df_one_hot.insert(len(df_one_hot.columns), 'Risk', risk)
```

```python
# Convert boolean columns to integers (1 for True, 0 for False)
boolean_columns = df_one_hot.select_dtypes(include=['bool']).columns.tolist()

for column in boolean_columns:
    df_one_hot[column] = df_one_hot[column].astype(int)

# Display the transformed DataFrame
print("\nDataFrame with Boolean Columns Converted to Integers:")
print(df_one_hot)
```

```python
# Output df_cleaned to a CSV file
df_one_hot.to_csv('C:/Users/Samsung/Downloads/cleaned_german_credit_data_updated.csv', index=False)
```
